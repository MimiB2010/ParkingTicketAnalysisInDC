### We compiled number of total tickets issued by month and year from OpenData.DC.gov and used the data to plot the trend over the years. 

```python
%matplotlib notebook
```

```python
# Import Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Create a Dataframe using Data from Dc open
data=pd.DataFrame(pd.read_csv("ParkingData/Five_Year_By_Month.csv"))
data
```

```python
x_axis=np.arange(0,len(data["Year"]))
y_axis=data["Total Violation"]//1000
plt.plot(x_axis,y_axis,color="lightcoral",linewidth=1, marker="o")

plt.ylabel("Tickets Issued (in thousands)")
plt.xlabel("5 Year")
plt.title("Tickets Issued from 2014 to 2018")
plt.xticks(np.arange(0,len(data["Year"]), 12),["January 2014","January 2015","January 2016","January 2017","January 2018"],rotation=40)
plt.savefig("Output/Five Year trend by Month.png",dpi=500, bbox_inches="tight")
plt.show()
```
