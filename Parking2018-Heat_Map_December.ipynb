```python
# Import Dependencies
import pandas as pd
import requests
import json
import gmaps
from config import gkey
gmaps.configure(api_key=gkey)
```

```python
   # Create a reference the CSV file desired
csv_path = "Parking_Violations_Issued_in_December_2018.csv"

# Read the CSV into a Pandas DataFrame
Parking2018 = pd.read_csv(csv_path)

# Print the first five rows of data to the screen
Parking2018 = Parking2018.drop(columns=["ROWID_","DAY_OF_WEEK",
                          "BODY_STYLE","HOLIDAY",
                          "WEEK_OF_YEAR","MONTH_OF_YEAR","ISSUE_TIME"])
Parking2018.head()
```

```python
Parking2018 = Parking2018.round({"X": 3, "Y":3})
Parking2018.head()
```

```python
Parking2018_Grouped = Parking2018.groupby(["X","Y"]).count()
Parking2018_Grouped = Parking2018_Grouped.reset_index()

# Store latitude and longitude in locations
locations = Parking2018_Grouped[["Y", "X"]]

# Fill NaN values and convert to float
count = Parking2018_Grouped["OBJECTID"].astype(float)
Parking2018_Grouped
```

```python
fig = gmaps.figure(map_type="HYBRID")

# Create heat layer
heat_layer = gmaps.heatmap_layer(locations, weights=count, dissipating=False, max_intensity=300, point_radius=.0015)


# # Add layer
fig.add_layer(heat_layer)

# Display figure
fig
```
