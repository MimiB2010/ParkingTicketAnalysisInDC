```python
# Import Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#import March Data
March2015=pd.DataFrame(pd.read_csv("ParkingData/MarMOMData/Parking_Violations_in_March_2015.csv"))
March2016=pd.DataFrame(pd.read_csv("ParkingData/MarMOMData/Parking_Violations_Issued_in_March_2016.csv"))
March2017=pd.DataFrame(pd.read_csv("ParkingData/MarMOMData/Parking_Violations_Issued_in_March_2017.csv"))
March2018=pd.DataFrame(pd.read_csv("ParkingData/MarMOMData/Parking_Violations_Issued_in_March_2018.csv"))

#import June Data
June2014=pd.DataFrame(pd.read_csv("ParkingData/JunMOMData/Parking_Violations_Issued_in_June_2014.csv"))
June2015=pd.DataFrame(pd.read_csv("ParkingData/JunMOMData/Parking_Violations_Issued_in_June_2015.csv"))
June2016=pd.DataFrame(pd.read_csv("ParkingData/JunMOMData/Parking_Violations_Issued_in_June_2016.csv"))
June2017=pd.DataFrame(pd.read_csv("ParkingData/JunMOMData/Parking_Violations_Issued_in_June_2017.csv"))
June2018=pd.DataFrame(pd.read_csv("ParkingData/JunMOMData/Parking_Violations_Issued_in_June_2018.csv"))

#import September Data
Sept2014=pd.DataFrame(pd.read_csv("ParkingData/SepMOMData/Parking_Violations_Issued_in_September_2014.csv"))
Sept2015=pd.DataFrame(pd.read_csv("ParkingData/SepMOMData/Parking_Violations_Issued_in_September_2015.csv"))
Sept2016=pd.DataFrame(pd.read_csv("ParkingData/SepMOMData/Parking_Violations_Issued_in_September_2016.csv"))
Sept2017=pd.DataFrame(pd.read_csv("ParkingData/SepMOMData/Parking_Violations_Issued_in_September_2017.csv"))
Sept2018=pd.DataFrame(pd.read_csv("ParkingData/SepMOMData/Parking_Violations_Issued_in_September_2018.csv"))

#import December Data
Dec2014=pd.DataFrame(pd.read_csv("ParkingData/DecMOMData/Parking_Violations_Issued_in_December_2014.csv"))
Dec2015=pd.DataFrame(pd.read_csv("ParkingData/DecMOMData/Parking_Violations_Issued_in_December_2015.csv"))
Dec2016=pd.DataFrame(pd.read_csv("ParkingData/DecMOMData/Parking_Violations_Issued_in_December_2016.csv"))
Dec2017=pd.DataFrame(pd.read_csv("ParkingData/DecMOMData/Parking_Violations_Issued_in_December_2017.csv"))
Dec2018=pd.DataFrame(pd.read_csv("ParkingData/DecMOMData/Parking_Violations_Issued_in_December_2018.csv"))
```

```python
#Create 2014 Graph 
fourteenX=[0,len(June2014),len(Sept2014),len(Dec2014)]

#plt.bar(fourteenX, y, color='lightcoral', alpha=0.5, align="center")

#Create 2015 Graph 
fifteenX=[len(March2015),len(June2015),len(Sept2015),len(Dec2015)]
#plt.bar(fifteenXs,y, color='skyblue', alpha=0.5, align="center")

#Create 2016 Graph
sixteenX=[len(March2016),len(June2016),len(Sept2016),len(Dec2016)]
#plt.bar(sixteenX, y, color='lightgold', alpha=0.5, align="center")

#Create 2017 Graph 
seventeenX=[len(March2017),len(June2017),len(Sept2017),len(Dec2017)]
#plt.bar(seventeenX, y, color='lightpurple', alpha=0.5, align="center")

#Create 2018 Graph 
eighteenX=[len(March2018),len(June2018),len(Sept2018),len(Dec2018)]
#plt.bar(eighteenX, y, color='lightgreen', alpha=0.5, align="center")

values=pd.DataFrame({"2014":fourteenX,"2015":fifteenX,"2016":sixteenX,"2017":seventeenX,"2018":eighteenX})
values.plot.bar(rot=0)

months=["March","June","September","December"]

plt.xticks(values.index,months)

plt.title("Month to Month Comparisons")
plt.ylabel("Number of Violations")
plt.xlabel("Months")
```
